```python
# Define the fertilizer data for each split dose
doses = [
    {"amount": 50, "type": "Urea"},
    {"amount": 40, "type": "Urea"},
    {"amount": 30, "type": "DAP"},
    {"amount": 25, "type": "KCl"}
]

total_days = 80
interval = 20
cumulative_usage = 0

print(f"{'Day':<10} | {'Dose #':<10} | {'Type':<10} | {'Amount (kg)':<15} | {'Cumulative (kg)':<15}")
print("-" * 70)

# Loop through the schedule
# range(start, stop, step) handles the 20-day intervals perfectly
for i, day in enumerate(range(0, total_days, interval)):
    dose_info = doses[i]
    amount = dose_info["amount"]
    fert_type = dose_info["type"]

    cumulative_usage += amount

    print(f"Day {day:<6} | Dose {i+1:<5} | {fert_type:<10} | {amount:<13} | {cumulative_usage:<15}")

print("-" * 70)
print(f"Total Fertilizer Applied over {total_days} days: {cumulative_usage} kg/hectare")
```
