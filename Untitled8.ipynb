```python
# Define the fertilizer data for each split dose
doses = [
    {"amount": 50, "type": "Urea"},
    {"amount": 40, "type": "Urea"},
    {"amount": 30, "type": "DAP"},
    {"amount": 25, "type": "KCl"}
]

total_days = 80
interval = 20
cumulative_usage = 0

print(f"{'Day':<10} | {'Dose #':<10} | {'Type':<10} | {'Amount (kg)':<15} | {'Cumulative (kg)':<15}")
print("-" * 70)

# Loop through the schedule
# range(start, stop, step) handles the 20-day intervals perfectly
for i, day in enumerate(range(0, total_days, interval)):
    dose_info = doses[i]
    amount = dose_info["amount"]
    fert_type = dose_info["type"]

    cumulative_usage += amount

    print(f"Day {day:<6} | Dose {i+1:<5} | {fert_type:<10} | {amount:<13} | {cumulative_usage:<15}")

print("-" * 70)
print(f"Total Fertilizer Applied over {total_days} days: {cumulative_usage} kg/hectare")
```

### Using NumPy Arrays for Fertilizer Data

We can represent the fertilizer amounts as a NumPy array, which allows for vectorized operations like calculating the cumulative sum more efficiently.

```python
import numpy as np

# Original doses data
doses_data = [
    {"amount": 50, "type": "Urea"},
    {"amount": 40, "type": "Urea"},
    {"amount": 30, "type": "DAP"},
    {"amount": 25, "type": "KCl"}
]

# Extract amounts into a NumPy array
amounts_np = np.array([d['amount'] for d in doses_data])
# Extract types into a list (NumPy arrays are generally for homogeneous data types)
types_list = [d['type'] for d in doses_data]

# Calculate cumulative usage using NumPy's cumsum function
cumulative_usage_np = np.cumsum(amounts_np)

total_days = 80
interval = 20

print(f"{'Day':<10} | {'Dose #':<10} | {'Type':<10} | {'Amount (kg)':<15} | {'Cumulative (kg)':<15}")
print("-" * 70)

# Loop through the schedule, using NumPy arrays
for i, day in enumerate(range(0, total_days, interval)):
    amount = amounts_np[i]
    fert_type = types_list[i]
    cumulative = cumulative_usage_np[i]

    print(f"Day {day:<6} | Dose {i+1:<5} | {fert_type:<10} | {amount:<13} | {cumulative:<15}")

print("-" * 70)
print(f"Total Fertilizer Applied over {total_days} days: {cumulative_usage_np[-1]} kg/hectare")
```
