```python
import numpy as np
import matplotlib.pyplot as plt

# New Sample data for a similar example
np.random.seed(43) # Different seed for a new dataset
X_new = 3 * np.random.rand(100, 1)
y_new = 2 + 5 * X_new + np.random.randn(100, 1) * 0.2 # Different coefficients and noise

# Add bias term
X_new_b = np.c_[np.ones((100, 1)), X_new]

# Compute parameters θ = (XᵀX)⁻¹ Xᵀy (manual calculation)
theta_new = np.linalg.inv(X_new_b.T @ X_new_b) @ X_new_b.T @ y_new

print(f"New Intercept: {theta_new[0][0]:.2f}, New Slope: {theta_new[1][0]:.2f}")

# Plot
plt.scatter(X_new, y_new, label='New Data', alpha=0.7)
plt.plot(X_new, X_new_b @ theta_new, color='green', label='New Fitted Line')
plt.title('Linear Regression: New Example')
plt.xlabel('X_new')
plt.ylabel('y_new')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# -----------------------------
# 1. Generate Example Data
# -----------------------------
# For demonstration, we'll create synthetic humidity (%) and temperature (°C) data
np.random.seed(42)  # For reproducibility
humidity = np.random.uniform(30, 90, 50)  # Humidity between 30% and 90%
temperature = 35 - 0.2 * humidity + np.random.normal(0, 1.5, 50)
# Example: Higher humidity → slightly lower temperature

# Convert to DataFrame for clarity
df = pd.DataFrame({
    'Humidity': humidity,
    'Temperature': temperature
})

# -----------------------------
# 2. Prepare Data for Model
# -----------------------------
X = df[['Humidity']]  # Feature must be 2D
y = df['Temperature']  # Target

# -----------------------------
# 3. Train Linear Regression Model
# -----------------------------
model = LinearRegression()
model.fit(X, y)

# -----------------------------
# 4. Predictions
# -----------------------------
y_pred = model.predict(X)

# -----------------------------
# 5. Model Evaluation
# -----------------------------
print("Intercept:", model.intercept_)
print("Slope (Coefficient):", model.coef_[0])
print("Mean Squared Error:", mean_squared_error(y, y_pred))
print("R² Score:", r2_score(y, y_pred))

# -----------------------------
# 6. Visualization
# -----------------------------
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Humidity (%)')
plt.ylabel('Temperature (°C)')
plt.title('Linear Regression: Temperature vs Humidity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# New synthetic data for another example
np.random.seed(2023) # Another different seed
X_example = 5 * np.random.rand(100, 1)
y_example = 10 + 2 * X_example + np.random.randn(100, 1) * 0.5 # New coefficients and noise

# Create a Linear Regression model
model_example = LinearRegression()
model_example.fit(X_example, y_example)

# Make predictions
y_pred_example = model_example.predict(X_example)

# Evaluate the model
print(f"Intercept (Example): {model_example.intercept_.item():.2f}")
print(f"Slope (Example): {model_example.coef_[0].item():.2f}")
print(f"Mean Squared Error (Example): {mean_squared_error(y_example, y_pred_example):.2f}")
print(f"R-squared (Example): {r2_score(y_example, y_pred_example):.2f}")

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X_example, y_example, color='blue', label='Generated Data', alpha=0.7)
plt.plot(X_example, y_pred_example, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression Example with Scikit-learn')
plt.xlabel('X_example')
plt.ylabel('y_example')
plt.legend()
plt.grid(True)
plt.show()
```
