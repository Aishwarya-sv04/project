```python
class Crop:
  def __init__(self, name, season, expected_yeild ):
    self.name = name
    self.season = season
    self.expected_yeild = expected_yeild

```

```python
rice = Crop('rice', 'kharif', 5.2)
wheat = Crop('wheat', 'rabi', 4.5)

print(rice.name, rice.season, rice.expected_yeild)
print(wheat.name, wheat.season, wheat.expected_yeild)
```
